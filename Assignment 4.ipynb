#### 8. Details of Datasets from UCI machine learning repositories.
Url = https://archive.ics.uci.edu/ You have to find the following details:
    
A) Dataset name
B) Data type
C) Task
D) Attribute type
E) No of instances
F) No of attribute G) Year 

```python
driver = webdriver.Chrome()
```

```python
driver.get('https://archive.ics.uci.edu/')
```

```python
select = driver.find_element(By.XPATH,'//a[@class="btn-primary btn"]')
select.click()
```

```python
Dataset = []
Type = []
Task = []
Attribute = []
Instances = []
No_Attribute = []
Year = []
```

```python
db_url = []
start = 0
end = 6
for page in range(start,end): 
    url=driver.find_elements(By.XPATH,'//a[@class="link-hover link text-xl font-semibold"]')
    for i in url:    
        db_url.append(i.get_attribute("href"))
    nxt_button=driver.find_element(By.XPATH,'//div[@class="btn-group"]/button[2]')
    nxt_button.click()
```

```python
len(db_url)
```

```python
db_details=driver.find_element(By.XPATH,'//div[@class="relative col-span-8 sm:col-span-7"]')
db_details.click()
```

```python
for url in db_url:
    driver.get(url)
    
    try:
        ds = driver.find_elements(By.XPATH,'//h1[@class="text-3xl font-semibold text-primary-content"]')
        for i in ds:
            dataset = i.text
            Dataset.append(dataset)
    except NoSuchElementException:
        Dataset.append('-')
    
    try:
        tpe = driver.find_elements(By.XPATH,'//div[@class="relative flex flex-col gap-4 bg-base-100 p-4 shadow"]/div[2]/div[1]/p')
        for i in tpe:
            dstype = i.text
            Type.append(dstype)
    except NoSuchElementException:
        Type.append('-')
    
    try:
        tsk = driver.find_elements(By.XPATH,'//div[@class="relative flex flex-col gap-4 bg-base-100 p-4 shadow"]/div[2]/div[3]/p')
        for i in tsk:
            task = i.text
            Task.append(task)
    except NoSuchElementException:
        Task.append('-')
    try:
        att = driver.find_elements(By.XPATH,'//div[@class="relative flex flex-col gap-4 bg-base-100 p-4 shadow"]/div[2]/div[4]/p')
        for i in att:
            attribute = i.text
            Attribute.append(attribute)
    except NoSuchElementException:
        Attribute.append('-')
        
    try:
        inst = driver.find_elements(By.XPATH,'//div[@class="relative flex flex-col gap-4 bg-base-100 p-4 shadow"]/div[2]/div[5]/p')
        for i in inst:
            instance = i.text
            Instances.append(instance)
    except NoSuchElementException:
        Instances.append('-')
        
    try:
        noat = driver.find_elements(By.XPATH,'//div[@class="relative flex flex-col gap-4 bg-base-100 p-4 shadow"]/div[2]/div[6]/p')
        for i in noat:
            no_att = i.text
            No_Attribute.append(no_att)
    except NoSuchElementException:
        No_Attribute.append('-')
    
    try:
        yr = driver.find_elements(By.XPATH,'//h2[@class="text-sm text-primary-content"]')
        for i in yr:
            year = i.text
            Year.append(year)
    except NoSuchElementException:
        Year.append('-')
    
```

```python
print(len(Dataset),len(Type),len(Task),len(Attribute),len(Instances),len(No_Attribute),len(Year))
```

```python
import pandas as pd
df = pd.DataFrame({'Dataset name':Dataset,'Data type':Type,'Task':Task,'Attribute type':Attribute,'No of instances':Instances,'No of attributes':No_Attribute,'Year':Year})
df
```
